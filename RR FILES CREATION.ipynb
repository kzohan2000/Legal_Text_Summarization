```python
import json
from sklearn.feature_extraction.text import TfidfVectorizer

def compare_json_files(file_path1, file_pah2):
    with open(file_path1, 'r') as f:
        issue_data = json.load(f)
    issue_keys = issue_data["ISSUE"][0].keys()  
    with open(file_pah2, 'r') as f:
        fac_data = json.load(f)

    similarities = {}
    for issue_key in issue_keys:
        similarities[issue_key] = []
        for fac_segment in fac_data["FAC"]:
            if issue_key in fac_segment:
                other_key = issue_key
                other_text = fac_segment[issue_key]
                analysis_text = issue_data["ISSUE"][0][issue_key]
                vectorizer = TfidfVectorizer()
                issue_vector = vectorizer.fit_transform([analysis_text])
                fac_vector = vectorizer.transform([other_text])
                similarity = issue_vector.dot(fac_vector.T)[0, 0]
                similarities[issue_key].append((other_key, similarity))
    return similarities

file_path1 = "dict/ISSUE.json"  
file_path2 = "dict/FAC.json" 
similarities = compare_json_files(file_path1, file_path2)
print(f"Similarities between {file_path1} and {file_path2}:")
for key, comparisons in similarities.items():
    print(f"- Key: {key}")
    for other_key, similarity in comparisons:
        print(f"  - Key: {other_key}, Similarity: {similarity:.4f}")
```
